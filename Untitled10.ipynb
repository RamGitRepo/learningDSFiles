```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bloodsugar_men= [113,85,90,149,88]
bloodsugar_women= [67,98,89,120,133]
plt.xlabel("Sugar Range")
plt.ylabel("Total Person")
plt.title("Blood Sugar Analysis")
plt.hist([bloodsugar_men,bloodsugar_women],bins=[90,100,125,150],rwidth=0.75,color=["green","orange"],
         label=["Men","Women"])
plt.legend()
```
