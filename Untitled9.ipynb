```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ["TCS","Infosys","Genpact"]
revenue = [90,136,89]
profit = [40,2,34]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.yticks(xpos, company)
plt.ylabel("Revnue")
plt.xlabel("Indian Stocks")
plt.title("Share Market")
plt.bar(xpos-0.2,revenue,width=0.4,color="orange",label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,color="green",label="Profit")
plt.legend()
```

```python
plt.yticks(xpos, company)
plt.barh(xpos-0.2,revenue,color="orange",label="Revenue")
plt.barh(xpos+0.2,profit,color="green",label="Profit")
plt.legend()
```
